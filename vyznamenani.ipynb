- Řád Bílého lva je nejvyšší státní vyznamenání České republiky. Původně byl udělován osobám, které se zvlášť vynikajícím způsobem zasloužily o Československo, a to od roku 1922. V letech 1922–1990 se tento řád oficiálně nazýval „Československý řád Bílého lva“. 
    - _Řád Bílého lva propůjčuje nebo uděluje prezident republiky a) občanům České republiky za zvláště vynikající zásluhy o stát, zejména v oblasti politiky, správy státu, rozvoje hospodářství, vědy, techniky, kultury, umění, školství, za mimořádné zásluhy o obranu a bezpečnost státu, za vynikající velitelskou a bojovou činnost, za vědeckou a odbornou práci vojenského a branného charakteru a za významné proslavení vlasti v zahraničí, b) těm, kteří nejsou občany České republiky, za zvláště vynikající zásluhy ve prospěch České republiky._

```python
with open("zenskajmena.txt", 'r', encoding="utf-8") as zenskajmena:
    zenskajmena = zenskajmena.read().splitlines()
    zeny = []
    for line in zenskajmena:
        zeny.append(line + " ")
```

## Vícenásobně vyznamenaní

```python
vicenasobni_a = pd.DataFrame(metaly.groupby("jmeno")["vyznamenani"].nunique().nlargest(45))
vicenasobni_a = vicenasobni_a[vicenasobni_a["vyznamenani"] > 1]
vicenasobni_a
```

```python
vicenasobni = pd.DataFrame(metaly.groupby("jmeno")["rok"].nunique().nlargest(45))
vicenasobni = vicenasobni[vicenasobni["rok"] > 1]

poprve = pd.Series(metaly.groupby("jmeno")["rok"].first(), name="poprve")
prvni_prezident = pd.Series(metaly.groupby("jmeno")["prezident"].first(), name="prvni_prezident")
podruhe = pd.Series(metaly.groupby("jmeno")["rok"].last(), name="podruhe")
druhy_prezident = pd.Series(metaly.groupby("jmeno")["prezident"].last(), name="druhy_prezident")
prvni_metal = pd.Series(metaly.groupby("jmeno")["vyznamenani"].first(), name="prvni_metal")
druhy_metal = pd.Series(metaly.groupby("jmeno")["vyznamenani"].last(), name="druhy_metal")
prvni_rank = pd.Series(metaly.groupby("jmeno")["rank"].first(), name="prvni_rank")
druhy_rank = pd.Series(metaly.groupby("jmeno")["rank"].last(), name="druhy_rank")

vicenasobni = vicenasobni.merge(poprve, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(podruhe, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(prvni_prezident, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(druhy_prezident, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(prvni_metal, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(druhy_metal, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(prvni_rank, on="jmeno", how="left")
vicenasobni = vicenasobni.merge(druhy_rank, on="jmeno", how="left")

vicenasobni["rozestup"] = vicenasobni["podruhe"] - vicenasobni["poprve"]

vicenasobni.sort_values(by="rozestup")
```

## Syntéza

```python
synt_zeny = pd.Series(metaly[(metaly["stat"] == "cr") & (metaly["pohlavi"] == "zena")].groupby("prezident")["id"].nunique() / metaly[metaly["stat"] == "cr"].groupby("prezident")["id"].nunique(), name="zeny")

synt_prvnitridy = pd.Series(metaly[(metaly["stat"] == "cr") & (metaly["rank"] == 1)].groupby("prezident")["id"].nunique() / metaly[metaly["stat"] == "cr"].groupby("prezident")["id"].nunique(), name="prvni tridy a stupne")

synt_inmemoriam = pd.Series(metaly[(metaly["stat"] == "cr") & (metaly["inmemoriam"] == True)].groupby("prezident")["id"].nunique() / metaly[metaly["stat"] == "cr"].groupby("prezident")["id"].nunique(), name="in memoriam")

synt_vyznamni1000_zivi = pd.Series(lvi_extended[(lvi_extended["stat"] == "cr") & (lvi_extended["zebricek"] < 1000) & (lvi_extended["inmemoriam"] == False)].groupby("prezident")["id"].nunique() / lvi_extended[(lvi_extended["stat"] == "cr")  & (lvi_extended["inmemoriam"] == False)].groupby("prezident")["id"].nunique(), name="tisicovka nejvyznamejsich lidi, bez in memoriam")

synt_vyznamni100000_zivi = pd.Series(lvi_extended[(lvi_extended["stat"] == "cr") & (lvi_extended["zebricek"] < 100000) & (lvi_extended["inmemoriam"] == False)].groupby("prezident")["id"].nunique() / lvi_extended[(lvi_extended["stat"] == "cr")  & (lvi_extended["inmemoriam"] == False)].groupby("prezident")["id"].nunique(), name="sto tisic nejvyznamejsich lidi, bez in memoriam")
```

```python
synt2_inmemoriam_rekord = pd.Series(lvi_extended[(lvi_extended["stat"] == "cr") & (lvi_extended["inmemoriam"] == True)].groupby("prezident")["posmrti"].max(), name="in memoriam ŘBL nejdele po smrti")
synt2_inmemoriam_median = pd.Series(lvi_extended[(lvi_extended["stat"] == "cr") & (lvi_extended["inmemoriam"] == True)].groupby("prezident")["posmrti"].median(), name="in memoriam ŘBL medián")
```

```python
synteza = pd.DataFrame(index=["Havel","Klaus","Zeman"])
synteza = synteza.merge(synt_zeny.to_frame(), left_index=True, right_index=True)
synteza = synteza.merge(synt_prvnitridy.to_frame(), left_index=True, right_index=True)
synteza = synteza.merge(synt_inmemoriam.to_frame(), left_index=True, right_index=True)
synteza = synteza.merge(synt_vyznamni1000_zivi.to_frame(), left_index=True, right_index=True)
```

```python
synteza2 = pd.DataFrame(index=["Havel","Klaus","Zeman"])
synteza2 = synteza2.merge(synt2_inmemoriam_median.to_frame(), left_index=True, right_index=True)
synteza2 = synteza2.merge(synt2_inmemoriam_rekord.to_frame(), left_index=True, right_index=True)
```

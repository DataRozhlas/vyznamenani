- Řád Bílého lva je nejvyšší státní vyznamenání České republiky. Původně byl udělován osobám, které se zvlášť vynikajícím způsobem zasloužily o Československo, a to od roku 1922. V letech 1922–1990 se tento řád oficiálně nazýval „Československý řád Bílého lva“. 
    - _Řád Bílého lva propůjčuje nebo uděluje prezident republiky a) občanům České republiky za zvláště vynikající zásluhy o stát, zejména v oblasti politiky, správy státu, rozvoje hospodářství, vědy, techniky, kultury, umění, školství, za mimořádné zásluhy o obranu a bezpečnost státu, za vynikající velitelskou a bojovou činnost, za vědeckou a odbornou práci vojenského a branného charakteru a za významné proslavení vlasti v zahraničí, b) těm, kteří nejsou občany České republiky, za zvláště vynikající zásluhy ve prospěch České republiky._
    - S řádovým řetězem se vyznamenání uděluje pouze hlavám států.
    - Českým státním občanům se řád neuděluje, ale propůjčuje.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cr_lev = pd.read_json("cr-rad-bileho-lva.json")
cr_masaryk = pd.read_json("cr-rad-t.-g.-masaryka.json")
cr_zasluhy = pd.read_json("cr-medaile-za-zasluhy.json")
cr_hrdinstvi = pd.read_json("cr-medaile-za-hrdinstvi.json")
csfr_lev = pd.read_json("csfr-rad-bileho-lva.json")
csfr_masaryk = pd.read_json("csfr-rad-t.-g.-masaryka.json")
csfr_stefanik = pd.read_json("csfr-rad-m.-r.-stefanika.json")
csfr_zasluhy = pd.read_json("csfr-medaile-za-zasluhy.json")
csfr_hrdinstvi = pd.read_json("csfr-medaile-za-hrdinstvi.json")
```

```python
Havel = [1990, 1991, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002]
Klaus = [2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012]
Zeman = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
Strasky = [1992]
```

```python
metaly = pd.concat([cr_lev, cr_masaryk, cr_zasluhy, cr_hrdinstvi, csfr_lev, csfr_masaryk, csfr_stefanik, csfr_hrdinstvi, csfr_zasluhy], axis=0, join='outer')
```

```python
with open("zenskajmena.txt", 'r', encoding="utf-8") as zenskajmena:
    zenskajmena = zenskajmena.read().splitlines()
    zeny = []
    for line in zenskajmena:
        zeny.append(line + " ")
```

```python
metaly.loc[metaly.jmeno.eq("Petra Kellner"),"jmeno"]="Petr Kellner"
metaly.loc[metaly.jmeno.eq("Hana"),"jmeno"]="Hana Hegerová"

metaly.loc[metaly.jmeno.eq("Hanuš"),"jmeno"]="Hanuš Holzer"
metaly.loc[metaly.jmeno.eq("Kare"),"jmeno"]="Karel Kryl"
metaly.loc[metaly.jmeno.eq("Karel"),"jmeno"]="Karel Hanzl"

metaly.loc[metaly.jmeno.eq("Alfred Duff Cooper"),"inmemoriam"]=True
metaly.loc[metaly.jmeno.eq("Lech Aleksander Kaczyński"),"inmemoriam"]=True
```

```python
metaly["pohlavi"] = "muz"
metaly.loc[metaly.jmeno.str.contains('|'.join(zeny)),"pohlavi"]="zena"
```

```python
metaly = metaly[metaly["rok"].notnull()]
metaly = metaly[metaly["jmeno"].notnull()]
```

```python
metaly.loc[metaly.rok.isin(Havel),"prezident"]="Havel"
metaly.loc[metaly.rok.isin(Klaus),"prezident"]="Klaus"
metaly.loc[metaly.rok.isin(Zeman),"prezident"]="Zeman"
metaly.loc[metaly.rok.isin(Strasky),"prezident"]="Stráský"
```
